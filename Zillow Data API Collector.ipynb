```python
import pandas as pd
from pathlib import Path # importing files
from pyzillow.pyzillow import ZillowWrapper, GetDeepSearchResults # Zillow information
```

```python
df_tra = pd.read_csv('Desktop/Zillow Data/df_travis.csv')
```

```python
df_tra
```

```python
# Temp lists to store data
home_type = []
year_built = []
home_size = []
bathrooms = []
bedrooms = []
last_sold_date = []
last_sold_price = []
latitude = []
longitude = []
zestimate_amount = []
address_ = []
zip_code = []
```

```python
#Creating a loop to download data
for i in range(31656, 41001):
    
    try:
        addres = df_tra.Address.iloc[i]
        zipcode = df_tra.zipcodes.iloc[i]
        zillow_data = ZillowWrapper('XXXXXXXXXXXX')
        deep_search_response = zillow_data.get_deep_search_results(addres, zipcode)
        result = GetDeepSearchResults(deep_search_response)

        
        home_type.append(result.home_type)
        year_built.append(result.year_built)
        home_size.append(result.home_size)
        bathrooms.append(result.bathrooms)
        bedrooms.append(result.bedrooms)
        last_sold_date.append(result.last_sold_date)
        last_sold_price.append(result.last_sold_price)
        latitude.append(result.latitude)
        longitude.append(result.longitude)
        zestimate_amount.append(result.zestimate_amount)
        address_.append(df_tra.Address.iloc[i])
        zip_code.append(df_tra.zipcodes.iloc[i])
    except:
        print(i)
```

```python
# Creating Data Frame 
df = pd.DataFrame(home_type, columns = ['home_type']) 
df['home_size'] = home_size
df['bathrooms'] = bathrooms
df['bedrooms'] = bedrooms
df['last_sold_date'] = 0
df['last_sold_price'] = 0
df['latitude'] = latitude
df['longitude'] = longitude
df['zestimate_amount'] = zestimate_amount
df['year_built'] = year_built
df['address'] = address_
df['zipcode'] = zip_code
```

```python
# writing a csv file
df.to_csv('df_31656_40980.csv')
```
