# Cleaning Travis County Dataset

First I need to remove buisnesses from my dataset

```python
df_tra['company']=df_tra['Owner Name'].apply(lambda x: 1 if 'LLC' in x else 0) # dummy variable
df_tra = df_tra[df_tra.company==0] 
df_tra['company_1']=df_tra['Owner Name'].apply(lambda x: 1 if any(i.isdigit() for i in x) else 0) # dummy variable
df_tra = df_tra[df_tra.company_1==0] 
```

```python
df_tra
```

Now will delete all "personal" type of property

```python
df_tra = df_tra[df_tra.Type=='Real'] 
del df_tra['company']
del df_tra['company_1']
del df_tra['Geographic ID']
del df_tra['Property ID']
del df_tra['Type']
del df_tra['Legal Description']
del df_tra['Owner Name']
del df_tra['Doing Business As']
```

```python
list_zips = []
for row_index, row  in df_tra['Property Address'].iteritems():
    
    zipcode = row.split()[-1][:5]
    list_zips.append(zipcode)
```

```python
df_tra['zipcodes'] = list_zips
```

```python
list_addresses = []
for row_index, row  in df_tra['Property Address'].iteritems():

    address = ' '.join(row.split()[:-1])
    list_addresses.append(address)
```

```python
df_tra['Address'] = list_addresses
```

```python
del df_tra['Property Address']
df_tra = df_tra.reset_index()
df_tra = df_tra.drop_duplicates(subset='Address', keep='first')
```
